```python
import chess.pgn

#games = open("lichess_db_standard_rated_2013-01.pgn", encoding="utf8", errors='ignore')
games2 = open("lichess_db_standard_rated_2017-02.pgn", encoding="utf8", errors='ignore')
```

```python
#game = chess.pgn.read_game(games)
#game
```

```python
result2 = {}
i = 0
while True:
    i += 1
    game = chess.pgn.read_game(games2)
    if game is None:
        break

    headers = dict(game.headers)
    headers["moves"] = game.board().variation_san(game.mainline_moves())

    result2["Game{}".format(i)] = headers
```

```python
import pandas as pd
df2 = pd.DataFrame(result2)
df_t2 = df2.transpose()
```
